```python
cmssw_pop=[]
popdb_pop=[]
diff_pop=[]
for l in open('corr_cmssw_popdb.txt'):
    sp=l.split()
    if len(l)<3: continue
    if 'AOD' not in sp[0]: continue
    cmssw_pop.append(0.1+float(sp[1]))
    popdb_pop.append(0.1+float(sp[2]))
    diff_pop.append((cmssw_pop[-1]-popdb_pop[-1])/(0.5*(popdb_pop[-1]+cmssw_pop[-1])))

%matplotlib inline
import pylab
pylab.rcParams['figure.figsize'] = 10, 6

pylab.figure(1)
ax = pylab.gca()
ax.scatter(cmssw_pop,popdb_pop)
pylab.xlabel('Files read in dataset (CMSSW popularity)',fontsize=15)
pylab.ylabel('Files read in dataset (PopDB popularity)',fontsize=15)
ax.set_xscale('log')
ax.set_yscale('log')
pylab.plot([1,100000],[1,100000],'r',linewidth=3)

pylab.figure(2)
pylab.hist(diff_pop,linewidth=3,bins=100)
pylab.ylabel('Datasets',fontsize=15)
pylab.xlabel('Files read CMSSW pop - PopDB pop (Diff/Average)',fontsize=15)
```
